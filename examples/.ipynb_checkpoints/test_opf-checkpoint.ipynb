```julia
using PowerModels

data = PowerModels.parse_file("/home/hoanganh/Desktop/math-topics/InterruptedRelax/codes/pglib-opf-master/sad/pglib_opf_case14_ieee__sad.m")

using DynamicPolynomials

include("../src/InterRelax.jl")
using .InterRelax

x,f,g,h,startpoint=InterRelax.get_POP_OPF(data);

#g=Vector{Polynomial{true, Float64}}([])

println("Number of variables: n=",length(x))
println("Number of inequality constraints: m=",length(g))
println("Number of equality constraints: l=",length(h))
```

```julia
include("../src/InterRelax.jl")
using .InterRelax


n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterRelax.get_info(x,f,g,h,sparse=true);
opt_val_local=InterRelax.POP_NLP(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,startpoint)
```

```julia
opt_val_local=3.2936e5
```

```julia
include("../src/InterRelax.jl")
using .InterRelax


n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterRelax.get_info(x,f,g,h,sparse=true);


d=Int64(maximum([sum(supp_f[:,i]) for i in 1:lmon_f]))+1
k=0
s=50

@time opt_val,opt_sol=InterRelax.RelaxSparse(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,
    lmon_f,supp_f,coe_f,dg,dh,k,s,d;assign="all",alg="MD",minimize=true,solver="Mosek",order=k)

println("gap = ",InterRelax.gap(opt_val,opt_val_local)*100)
```

```julia
using SparseArrays

include("../src/InterRelax.jl")
using .InterRelax

n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterRelax.get_info(x,f,g,h,sparse=true);

using TSSOS

k=1

#vars,pop=InterRelax.get_POP(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f)
vars=x
pop=[[f];[g;x];h]

@time opt_val,sol,data=TSSOS.cs_tssos_first(pop,vars,k,numeq=l,CS="MD",TS="block", merge=false,solver="Mosek");

println("gap = ",InterRelax.gap(opt_val,opt_val_local)*100)
```

```julia
@time opt_val,sol,data = cs_tssos_higher!(data; TS="block", merge=false, md=3, QUIET=false, solve=true,
solution=false, MomentOne=false);
if opt_val != nothing
    println("gap = ",InterRelax.gap(opt_val,opt_val_local)*100)
end
```
