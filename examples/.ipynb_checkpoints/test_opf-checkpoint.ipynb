```julia
using PowerModels

data = PowerModels.parse_file("/home/hoanganh/Desktop/math-topics/InterruptedRelax/codes/pglib-opf-master/sad/pglib_opf_case5_pjm__sad.m")

using DynamicPolynomials

include("../src/InterruptedRelax.jl")
using .InterruptedRelax

x,f,g,h,startpoint=InterruptedRelax.get_POP_OPF(data);

#g=Vector{Polynomial{true, Float64}}([])

println("Number of variable: n=",length(x))
println("Number of inequality constraints: m=",length(g))
println("Number of equality constraints: l=",length(h))

```

```julia
include("../src/InterruptedRelax.jl")
using .InterruptedRelax


n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterruptedRelax.get_info(x,f,g,h,sparse=true);


opt_val_local=InterruptedRelax.POP_NLP(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,startpoint)
```

```julia
include("../src/InterruptedRelax.jl")
using .InterruptedRelax


n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterruptedRelax.get_info(x,f,g,h,sparse=true);


d=Int64(maximum([sum(supp_f[:,i]) for i in 1:lmon_f]))+1

println("Degree: d=",d)


k=1

println("Relaxation order: k=",k)


s=50


println("Block size: s=",s)

@time opt_val=InterruptedRelax.RelaxSparse(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,
    lmon_f,supp_f,coe_f,dg,dh,k,s,d;assign="all",alg="MD",minimize=true,solver="Mosek")

println("gap = ",InterruptedRelax.gap(opt_val,opt_val_local)*100)
```

```julia
include("../src/InterruptedRelax.jl")
using .InterruptedRelax


n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterruptedRelax.get_info(x,f,g,h,sparse=true);


d=Int64(maximum([sum(supp_f[:,i]) for i in 1:lmon_f]))+1

println("Degree: d=",d)


k=0

println("Relaxation order: k=",k)


s=20


println("Block size: s=",s)

@time opt_val=InterruptedRelax.RelaxSparse_without_multiplier(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,
    lmon_f,supp_f,coe_f,dg,dh,s,d;assign="all",alg="MD",minimize=true,solver="Mosek",tol=1e-8)

println("gap = ",InterruptedRelax.gap(opt_val,opt_val_local)*100)
```

```julia
using SparseArrays

include("../src/InterruptedRelax.jl")
using .InterruptedRelax

n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f,dg,dh=InterruptedRelax.get_info(x,f,g,h,sparse=true);

using TSSOS

k=1

#vars,pop=InterruptedRelax.get_POP(n,m,l,lmon_g,supp_g,coe_g,lmon_h,supp_h,coe_h,lmon_f,supp_f,coe_f)
vars=x
pop=[[f];[g;x];h]

@time opt_val,sol,data=TSSOS.cs_tssos_first(pop,vars,k,numeq=l,CS="MD",TS="block", merge=false,solver="Mosek");

println("gap = ",InterruptedRelax.gap(opt_val,opt_val_local)*100)
```

```julia
@time opt_val,sol,data = cs_tssos_higher!(data; TS="block", merge=false, md=3, QUIET=false, solve=true,
solution=false, MomentOne=false);
if opt_val != nothing
    println("gap = ",InterruptedRelax.gap(opt_val,opt_val_local)*100)
end
```
